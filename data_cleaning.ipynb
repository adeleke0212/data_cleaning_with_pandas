```python
# Define a function to remove whitespace and use apply on the DF
# My code ran fine converting to datetime but in class didnt work untill white space removed

def remove_white_space(text):
    if type(text) == str:
        return text.strip()
    return text
```

```python
travels_data['Travel Start Date'] = travels_data['Travel Start Date'].apply(remove_white_space) #we dont have to call the function
```

```python
# Lets remove the initials
# Let's write a fucntion to split state and city from Destination
def extract_state(text):
    result = text.split(',')[1]
    return result

def extract_city(text):
    """Extract city"""
    result = text.split(',')[0]
    return result
```

```python
# Extract State and City
travels_data['State'] = travels_data['Destination'].apply(extract_state)
travels_data['City'] = travels_data['Destination'].apply(extract_city)
```
